```python
from matplotlib import pyplot as plt

plt.style.use("fivethirtyeight")


plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()

# Colors:
# Blue = #008fd5
# Red = #fc4f30
# Yellow = #e5ae37
# Green = #6d904f
```

```python
#how it works
slices =[60,40]
plt.pie(slices)
```

```python
#With labels
slices =[60,40, 30, 20]
labels = ['a', 'b', 'c', 'd']
col = ['#008fd5', '#fc4f30', '#e5ae37', '#6d904f']
#wedge property to make separation more visible
plt.pie(slices, labels = labels,colors = col, wedgeprops= {'edgecolor': 'black'})

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030,
          20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java',
          'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):',
          'Ruby', 'Go', 'Assembly']

#Plotting
plt.style.use("fivethirtyeight")

plt.pie(slices, labels = labels)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
#pie charts make very crowded

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#Plotting
plt.style.use("fivethirtyeight")

# we can use explode to emphasis on particular data, float values with respect to data index
explode = (0.1, 0, 0, 0.1, 0)
#to make shadow but shodow = True
plt.pie(slices, labels = labels, explode = explode, shadow = True, startangle= 90)

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```

```python
#If we want to get percentage of data

# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']

#Plotting
plt.style.use("fivethirtyeight")

plt.pie(slices, labels = labels, autopct= '%1.1f%%')

plt.title("My Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
