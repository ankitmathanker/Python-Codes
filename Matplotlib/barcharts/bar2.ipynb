```python
#Whole code
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from operator import itemgetter

data = pd.read_csv('data.csv')
data.head()

#creating variables for columns
ids = data['Responder_id']
languages = data['LanguagesWorkedWith']

#for language columns in languages
prog_lang = []
for response in languages:
    prog_lang.append(response.split(';'))
    #prog_lang.extend(response.split(';'))
prog_lang

lang = Counter()
for item in prog_lang:
    lang.update(item)
lang #unsorted

from operator import itemgetter
#https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value
sorted_x = sorted(lang.items(), key = itemgetter(1), reverse = True)
sorted_x

language = []
popularity = []
for item in sorted_x:
    language.append(item[0])
    popularity.append(item[1])
#print(language,'\n\n',popularity)

#Plotting
plt.figure(figsize = (10,7))
#plt.style.use('fivethirtyeight')
#reversing it 
language.reverse()
popularity.reverse()
plt.barh(language, popularity)
plt.title('Most population language data')
plt.xlabel('Popularity of language')

#to fit graph properly
plt.tight_layout()
#plt.savefig('format2.png')
plt.show()
```
