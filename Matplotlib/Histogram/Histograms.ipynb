```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('fivethirtyeight')


data = pd.read_csv('data.csv')
ids = data['Responder_id']
ages = data['Age']

#plot
bins = np.arange(10, 110, 10)
#plt.hist(ages, bins, edgecolor = 'k')

#if the last few values are not properly visible use log = True
plt.hist(ages, bins, edgecolor = 'k', log = True)


#plotting the median line
median_age = 29
color = '#fc4f30'
#axv ax vertical line
plt.axvline(median_age, color = color, linewidth = 1, label = 'Age Median')
plt.legend()

plt.title('Ages of Respondents')
plt.xlabel('Ages')
plt.ylabel('Total Respondents')

plt.tight_layout()

plt.show()
```
