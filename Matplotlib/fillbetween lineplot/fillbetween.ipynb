```python
#filling only above
#using 'where'
#to avoid omitted small x section, using interpolate
#using another fill between to fill below color
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287
#above
plt.fill_between(ages, py_salaries, overall_median, where = (py_salaries > overall_median),
                 interpolate=True, color ='b', alpha = 0.25)
#below
plt.fill_between(ages, py_salaries, overall_median, where = (py_salaries <= overall_median),
                 interpolate=True, color ='r', alpha = 0.25)

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

Filling area between two different plts

```python
#also label the fill between
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')


#above
plt.fill_between(ages, py_salaries, js_salaries, where = (py_salaries > js_salaries),
                 interpolate=True, color ='b', alpha = 0.25, label = 'Py Above Js')
#below
plt.fill_between(ages, py_salaries, js_salaries, where = (py_salaries <= js_salaries),
                 interpolate=True, color ='r', alpha = 0.25, label = 'Py Below Js')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```
